```python
import pandas as pd
from datetime import datetime, timedelta

# 주차 계산 함수
def calculate_week(target_date, base_week, startMonth, startDate):
    # 문자열 날짜를 datetime 객체로 변환
    date = pd.to_datetime(target_date, format='%Y%m%d')
    
    # 기준 날짜 설정
    base_date = datetime(2023, startMonth, startDate)
    base_date_timestamp = pd.Timestamp(base_date)

    # 날짜 차이 계산
    delta_days = (date - base_date_timestamp).days

    # 기준 주차에서 날짜 차이를 주 단위로 변환
    week = base_week + delta_days // 7
    print( date , f'는 {base_date}로 부터 {week}주차 입니다')
    return week

# 테스트
print(calculate_week('20230928', 1, 9, 20))

# # for문을 사용하여 주차 데이터를 데이터프레임에 추가
# for data, farm_name in zip(datalist, list):
#     print(f"{farm_name} 농가")
#     dates = pd.to_datetime(data['date'])
#     weeks = calculate_week(dates, base_week, 10, 28)
#     data['weeks'] = weeks
```
