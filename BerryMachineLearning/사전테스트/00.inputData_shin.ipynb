### 주차 df 만들기

```python
from datetime import datetime, timedelta

base_date = datetime(2023, 10, 6)
base_week = 4

# 주차 계산 함수
def calculate_week(date, base_date, base_week):
    base_date_timestamp = pd.Timestamp(base_date)

    # 날짜 차이 계산
    delta_days = (date - base_date_timestamp).dt.days

    # 기준 주차에서 날짜 차이를 주 단위로 변환
    week = base_week + delta_days // 7
    return week


# for돌리면서 주차데이터 df에 추가
for data, list in zip(datalist, list):
    print(f"{list} 농가")
    dates = pd.to_datetime(data['date'])
    weeks = calculate_week(dates, base_date, base_week)
    data['weeks'] = weeks

```

### CO2 컬럼 정규화 하기


```python
dataB2
```

```python
from sklearn.preprocessing import MinMaxScaler
datalist2 = [dataB2, dataC2, dataD2, dataE2]

for data in datalist2:

    min_x = data['innerCO2'] - data['innerCO2'].min()
    min_max = data['innerCO2'].max() - data['innerCO2'].min()
    
    normalCO2 = min_x / min_max
    data['CO2'] = normalCO2
    

dataE2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

B = dataB2.iloc[:,1:].corr()
C = dataC2.iloc[:,1:].corr()
D = dataD2.iloc[:,1:].corr()
E = dataE2.iloc[:,1:].corr()

sns.heatmap(B, ax=axs[0,0], cmap='viridis', cbar=True)
axs[0, 0].set_title('B농가 히트맵')

sns.heatmap(C, ax=axs[0,1], cmap='viridis', cbar=True)
axs[0, 1].set_title('C농가 히트맵')

sns.heatmap(D, ax=axs[1,0], cmap='viridis', cbar=True)
axs[1, 0].set_title('D농가 히트맵')

sns.heatmap(E, ax=axs[1,1], cmap='viridis', cbar=True)
axs[1, 1].set_title('E농가 히트맵')

plt.tight_layout()

# 그래프 출력
plt.show()
```
