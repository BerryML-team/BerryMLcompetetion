```python
import matplotlib.pyplot as plt
import pandas as pd

# '기준년월'이 datetime 형식이 아니라면 변환해줍니다.
data29['기준연월'] = pd.to_datetime(data29['기준연월'], format='%Y%m')

# 특정 기관 선택
selected_agency = '서울 구로구 가족센터'

# 선택한 기관의 데이터만 필터링
filtered_data = data29[data29['아이돌봄기관명'] == selected_agency]

# 그래프 그리기
plt.figure(figsize=(15, 8))

columns_to_plot = ['서비스총돌보미수','시간제돌보미수', '영아종일제돌보미수', '종합형돌보미수', '질병아동돌보미수', '기관연계돌보미수']

for column in columns_to_plot:
    plt.plot(filtered_data['기준연월'], filtered_data[column], label=column, marker='o')

plt.title(f'{selected_agency} - 시간에 따른 돌보미 수 변화')
plt.xlabel('기준연월')
plt.ylabel('돌보미 수')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)

# x축 날짜 포맷 설정
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MonthLocator(interval=3))

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor
import joblib
import Function  # colored_text 함수를 사용하기 위해 import

def xgboost_regressor_predict(input_data, target_data, test_size=0.2, random_state=42):
    # 데이터 분할
    train_input, test_input, train_target, test_target = train_test_split(
        input_data, target_data, test_size=test_size, random_state=random_state
    )

    # XGBoost 모델 생성 및 학습
    xg_reg = XGBRegressor()
    multi_output_regressor_xg = MultiOutputRegressor(xg_reg)
    multi_output_regressor_xg.fit(train_input, train_target)

    # 모델 평가
    score = multi_output_regressor_xg.score(test_input, test_target)
    y_pred_xg = multi_output_regressor_xg.predict(test_input)
    mse = mean_squared_error(test_target, y_pred_xg)
    rmse = np.sqrt(mse)
    r2 = r2_score(test_target, y_pred_xg)

    print(Function.colored_text(f'XGB regression model score: {score}', 'yellow'))
    print(f'XGBoost regression model RMSE: {rmse:.2f}')
    print(f'XGBoost regression model R2 score: {r2:.2f}')

    # 교차 검증
    scores_cv = cross_val_score(multi_output_regressor_xg, train_input, train_target, 
                                scoring='neg_mean_squared_error', cv=10)
    rmse_cv = np.sqrt(-scores_cv)

    # 모델 저장
    joblib.dump(multi_output_regressor_xg, "XG_model")
    print(f'모델이 {"XG_model"} 이름으로 저장됨')
    print("\t", f"XGB cv score : {rmse_cv}")
    print("\t", f"XGB cv RMSE average : {rmse_cv.mean():.2f}")

    # 예측 결과와 실제 값 비교 (ModelTest.real_pred_compare 함수가 없으므로 주석 처리)
    # ModelTest.real_pred_compare(y_pred_xg, test_target, test_input)

    return multi_output_regressor_xg, y_pred_xg, test_target, test_input

# 사용 예시:
# 여기서 input_data는 특성(features)을, target_data는 목표 변수(target variables)를 포함하는 데이터프레임 또는 numpy 배열이어야 합니다.
# model, predictions, true_values, test_features = xgboost_regressor_predict(input_data, target_data)
```
