```python
import pandas as pd
import os

# 파일이 있는 디렉토리 경로
directory = "Data/시도별 인구수_1단위"

# 디렉토리 내의 모든 CSV 파일 목록 가져오기
file_list = [f for f in os.listdir(directory) if f.endswith('.csv')]

# 파일 목록을 정렬하여 순서대로 처리
file_list.sort()

# 첫 번째 파일 읽기 (기준 데이터프레임)
base_df = pd.read_csv(os.path.join(directory, file_list[0]), encoding='euc-kr')

# 나머지 파일들을 순차적으로 병합
for file_name in file_list[1:]:
    # 파일 읽기
    file_path = os.path.join(directory, file_name)
    df = pd.read_csv(file_path, encoding='euc-kr')
    
    # 파일명에서 연도와 월 추출
    year_month = file_name[:11].replace('_', '')
    suffix = f"_{year_month}"
    
    # 데이터프레임 병합
    base_df = pd.merge(base_df, df, on='행정구역', how='left', suffixes=('', suffix))

# 결과 확인
print(base_df.head())
print(base_df.columns)

# 결과 저장
base_df.to_csv("병합된_인구데이터_전체_1세별.csv", index=False, encoding='utf-8-sig')
```
