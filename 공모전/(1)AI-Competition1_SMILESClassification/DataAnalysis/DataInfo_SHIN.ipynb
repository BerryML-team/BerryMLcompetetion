```python
from rdkit import Chem
from rdkit.Chem.MolStandardize import rdMolStandardize
from rdkit.Chem import AllChem


smiles = trainData1['Smiles']
for smile in smiles:
    mol = Chem.MolFromSmiles(smile)

    # 표준화 클래스를 사용하여 분자 표준화
    # standardizer = rdMolStandardize.CleanupParameters()
    # standardizer.doCanonical = True
    # cleaned_mol = rdMolStandardize.Cleanup(mol)


    # # 이성질체 및 양성자화 상태의 표준화
    # parent_cleaned = rdMolStandardize.StandardizeSmiles(Chem.MolToSmiles(cleaned_mol))

    # print("Standardized SMILES:", parent_cleaned)

    # 모듈을 통해 표준화된 SMILES 생성
    standardized_smiles = Chem.MolToSmiles(mol, canonical=True)
    print("Standardized SMILES:", standardized_smiles)


```

```python
# smiles 문자열 토큰화
def tokenize_smiles(smiles):
    all_tokens = []
    for smile in smiles: 
        print("smile, ", smile )
        tokens = []
        i = 0
        while i < len(smile):
            if smile[i : i+2] in ["Cl", "Br"] :
                tokens.append(smile[i: i+2])
                i += 2
            else :
                tokens.append(smile[i])
                i += 1
        all_tokens.append(tokens)
    return all_tokens

tokens = tokenize_smiles(smiles)
for i, token_list in enumerate(tokens) :
    print(f"SMILES {i+1}: {token_list}")
```
