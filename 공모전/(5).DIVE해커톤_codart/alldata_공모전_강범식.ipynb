```python
# 누락된 지역 확인
missing_regions = [region for region in regions if not any(region in group for group in do.values())]

if missing_regions:
    print("누락된 지역:", missing_regions)
else:
    print("모든 지역이 do_mapping에 포함되어 있습니다.")
```

```python

# 도 분류 함수
def classify_region(region):
    for dod, region_list in do.items():
        if any(k in region for k in region_list):
            return dod
    return '기타'  # 해당 도가 없으면 '기타'로 분류

# 도 기준으로 분류
삼성['도'] = 삼성['가맹점소재지'].apply(classify_region)
```
